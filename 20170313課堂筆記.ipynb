```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([[-2,3],[-8,7],[1,-2],[2,-2]])
y=np.array([1,1,2,2])
```

```python
plt.scatter(x[:,0],x[:,1],s=50,c=y)
```

```python
from sklearn.svm import SVC
```

```python
clf=SVC()
```

```python
clf.fit(x,y)
```

```python
clf.predict([[-2,3]])
```

```python
clf.predict([[1,-2]])
```

```python
# x=-10..4,y=-4..8
np.arange(0,10,0.7)
```

```python
gd=np.array([(i,j)for i in np.arange(-10,4,0.6)for j in np.arange(-4,8,0.6)])
```

```python
plt.scatter(gd[:,0],gd[:,1],s=50,c=clf.predict(gd))
```

## K-Mean

```python
x=np.random.rand(300,2)
```

```python
plt.scatter(x[:,0],x[:,1],s=50)
```

```python
from sklearn.cluster import KMeans
```

```python
clf=KMeans(n_clusters=10)
```

```python
clf.fit(x)
```

```python
clf.labels_
```

```python
plt.scatter(x[:,0],x[:,1],s=50,c=clf.labels_)
```

```python
gd=np.array([(i,j)for i in np.arange(-0.2,1.2,0.05)for j in np.arange(-0.2,1.2,0.05)])
```

```python
plt.scatter(gd[:,0],gd[:,1],s=50,c=clf.predict(gd))
```
